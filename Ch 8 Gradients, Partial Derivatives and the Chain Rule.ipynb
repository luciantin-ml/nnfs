```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
```

```python
def fun(x, y):
    return 3*x**2 - y**2 + 2*y

def gradient_x(x):
    return 9*x**2

def gradient_y(y):
    return - 2 * y + 2


fig = plt.figure(figsize=(18,18), facecolor='white')
ax = fig.gca(projection='3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

```python
print(X.shape)
print(Y.shape)
print(Z.shape)

grad = np.array(np.gradient([X,Y,Z]))

print(grad.shape)
print(grad[0][0])



fig = plt.figure(figsize=(18,18), facecolor='white')
plt.plot(grad[0][0][:10],grad[0][2][:10])
plt.show()
```

```python
N = 10                                  # Use ten samples
x = np.linspace(0, np.pi*2, N)          # Equally spaced x values
y = np.sin(x)                           # Corresponding sine values
grads = np.gradient(y)

fig = plt.figure(figsize=(18,18), facecolor='white')
plt.plot(x, y, "-b", label="values")                  # the y values
plt.plot(x + 0.5, grads, "--r", label="gradients")    # the computed gradients
plt.legend()
plt.show()
```
